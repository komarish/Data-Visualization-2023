```python
# Seaborn is a library for solving visualization problems with a focus on working in the field of statistics.
# Seaborn is based on the Matplotlib library

# Seaborn — это библиотека для решения задач визуализации с ориентацией на работу в области статистики. 
# Базой seaborn является библиотека Matplotlib
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# For experiments, we will use the data loaded using the load_dataset() function.
# It takes the name of the desired data set as an argument.
# Для экспериментов будем использовать данные, загружаемые с помощью функции load_dataset(). 
# В качестве аргумента она принимает имя нужного набора данных.
```

```python
# Download the mpg dataset /Загрузим набор данных mpg:
    
mpg = sns.load_dataset("mpg")

# The load_dataset() function forms a dataset and returns it as a pandas.DataFrame object.
# Функция load_dataset() формирует набор данных и возвращает его в виде объекта pandas.DataFrame. 
# The mpg set contains information about the characteristics of a number of cars.
# В наборе mpg содержится информация о характеристиках ряда автомобилей. 

mpg.head()
```

```python
# Let's build the dependence of acceleration (acceleration) on the amount of horsepower (horsepower),
# while the point size will be determined by the number of cylinders 

# Построим зависимость ускорения (acceleration) от количества лошадиных сил (horsepower), 
# при этом размер точки будет определяться количеством цилиндров

sns.relplot(x="horsepower", y="acceleration", size="cylinders", data=mpg);
plt.show()
```

```python
# теперь машины с разными cylinders обозначаются разными цветами

g = sns.relplot(x="horsepower", y="acceleration", size="cylinders", data=mpg, hue='cylinders');
plt.show()
```

```python
# Building a Line Graph / Построение линейного графика

# To demonstrate how the line plotting function works, load the flights dataset,
# containing information about the number of passengers who used air transport

# Для демонстрации работы функции построения линейного графика загрузим набор данных flights, 
# содержащий информацию о количестве пассажиров, которые воспользовались авиатранспортом

flights = sns.load_dataset("flights")
flights.head()
```

```python
# Build the dependence of the number of passengers carried (passengers) on the year (year)
# Построим зависимость количества перевезённых пассажиров (passengers) от года (year)

sns.relplot(x="year", y="passengers", kind="line", legend="full", data=flights)
plt.show()
```

```python
# Consider working with categorical data on the example of the iris set (Fisher's Irises)
# Рассмотрим работу с категориальными данными на примере набора iris (Ирисы Фишера)

iris = sns.load_dataset("iris")
iris.head()
```

```python
# The downloaded dataset is the reference for study classification algorithms, it represents information about 150 copies
# of iris, 50 for each individual species: Iris bristly (setosa), Iris virginsky (virginica) and Iris multi-colored (versicolor).
# The following parameters are defined for each instance:
# • Length of outer perianth segment (sepal_length);
# • Width of outer perianth segment (sepal_width);
# • Length of inner perianth segment (petal_length);
# • Width of inner perianth segment (petal_width).
# To visualize this set, we use the catplot() function

# Загруженный набор данных является эталонным для изучения алгоритмов классификации, он представляет собой информацию о 150 
# экземплярах ириса по 50 на каждый отдельный вид: Ирис щетинистый (setosa), Ирис Виргинский (virginica) и 
# Ирис разноцветный (versicolor). Для каждого экземпляра определены следующие параметры:
# • Длина наружной доли околоцветника (sepal_length);
# • Ширина наружной доли околоцветника (sepal_width);
# • Длина внутренней доли околоцветника (petal_length);
# • Ширина внутренней доли околоцветника (petal_width).
# Для визуализации этого набора воспользуемся функцией catplot()

sns.catplot(x="species", y="sepal_length", kind="swarm", data=iris);
plt.show()
```

```python
# Visualizing relationships in data / Визуализация отношений в данных
# Line graph. lineplot() function / Линейный график. Функция lineplot()

# The lineplot() function is used to visualize relationships in data as a line graph. This type of graph is most often used 
# to visualize time series and relationships between variables that are continuous, such as temperature, pressure, cost, etc.

# Для визуализации отношений в данных в виде линейного графика применяется функция lineplot(). Такой тип графика чаще всего
# используется для визуализации временных рядов и зависимостей между переменными, имеющими непрерывный характер, например, 
# такими как температура, давление, стоимость и т.д.

# Set the chart style / Установим стиль оформления графиков

sns.set_style("darkgrid")

# Let's generate a set of random data and use the lineplot() function to show the relationship between them
# Сгенерируем набор случайных данных и воспользуемся функцией lineplot() для отображения зависимости между ними

np.random.seed(123)
x = [i for i in range(10)]
y = np.random.randint(10, size=len(x))
sns.lineplot(x, y)
plt.show()
```

```python
#изменили толщину линии

sns.lineplot(x, y, linewidth = 5)
plt.show()
```

```python
# In practice, data for analysis is most often prepared in the form of Series or DataFrame structures of the pandas library.
# Let's create some more datasets

# На практике данные для анализа чаще всего подготавливаются в виде структур Series или DataFrame библиотеки pandas. 
# Создадим ещё несколько наборов данных

np.random.seed(123)
sample = [i for i in range(10)]
y = np.random.randint(10, size=len(sample))
z = np.random.randint(4, size=len(sample))
data = [sample, y, z]
```

```python
# Merge them into one DataFrame / Объединим их в один DataFrame
df = pd.DataFrame({'sample':sample, 'y_val': y, 'z_val': z})
```

```python
df.head(10)
```

```python
# Display the dependence of y_val on sample from the df set using the lineplot() function
# Отобразим зависимость y_val от sample из набора df с помощью функции lineplot()

sns.lineplot(x='sample', y='y_val', data=df)
plt.show()
```

```python
# изменили цвет

sns.lineplot(x='sample', y='y_val', data=df,color='red')
plt.show()
```

```python
# If you need to display all elements from the DataFrame set, then you can pass it to lineplot () through the data parameter,
# without specifying x and y

# Если необходимо отобразить все элементы из DataFrame набора, то можно его передать в lineplot() через параметр data, 
# при этом не задавая x и y

sns.lineplot(data=df)
plt.show()
```

```python
# изменили маркеры

sns.lineplot(data=df, marker= 'v', markersize=10)
plt.show()
```

```python
# Display of mathematical expectation and confidence intervals
# Отображение математического ожидания и доверительных интервалов

# Let's plot the dependence of val_y on val_z. JNote: the same value of val_z (e.g. 0) will match several val_y values 
# (in our case, these are 1 and 9).
# Построим график зависимости val_y от val_z. JОбратите внимание: одному и тому же значению val_z (например, 0) 
# будет соответствовать несколько значений val_y (в нашем случае — это 1 и 9). 

sns.lineplot(x='z_val', y='y_val', data=df)
plt.show()
```

```python
# изменили значение доверительного интервала

sns.lineplot(x='z_val', y='y_val', data=df, ci=70)
plt.show()
```

```python
# In this case, not the values of the parameter y, but its mathematical expectation will be plotted along the ordinate axis. 
# For example, for z_val=0, y_val=(1+9)/2=5, which is what we see on the chart. The lighter area around the line is 
# the 95% confidence interval, if you want to display the standard deviation instead, set the ci parameter of the 
# lineplot() function to 'sd'

# В этом случае по оси ординат будут откладываться не значения параметра y, а его математическое ожидание. Например для z_val=0,
# y_val=(1+9)/2=5, что мы и видим на графике. Более светлая область вокруг линии — это 95% доверительный интервал, 
# если вместо него необходимо отобразить стандартное отклонение, то следует присвоить параметру ci функции lineplot() 
# значение 'sd'

sns.lineplot(x='z_val', y='y_val', ci='sd', data=df)
plt.show()
```

```python
# добавили количество бутстрап-выборок, используемых для вычисления доверительного интервала

sns.lineplot(x='z_val', y='y_val', ci='sd', n_boot=100, data=df)
plt.show()
```

```python
# The display of the confidence interval (or standard deviation) can be turned off by passing None to ci
# Отображение доверительного интервала (или стандартного отклонения) можно отключить, передав в ci значение None

sns.lineplot(x='z_val', y='y_val', ci=None, data=df)
plt.show()
```

```python
# изменили цвет

sns.lineplot(x='z_val', y='y_val', ci=None, data=df, color="#5A9")
plt.show()
```

```python
# Let's return to the example illustrating the dependence of passenger traffic on the year from the flights dataset
# Вернемся к примеру, иллюстрирующему зависимость пассажиропотока от года из набора данных flights

flights = sns.load_dataset("flights")
sns.lineplot(x='year', y='passengers', data=flights)
plt.show()
```

```python
# To control how the confidence interval is displayed (solid fill, labels), the err_style parameter is used, which can take the 
# values 'band', 'bars' or None. The 'band' value defines a solid fill, as shown in the figure above. 
# The value 'bars' displays the confidence interval as bars
# Для управления видом отображения доверительного интервала (сплошная заливка, метки) используется параметр err_style, который
# может принимать значения 'band', 'bars' или None. Значение 'band' определяет сплошную заливку, как это представлено на рисунке выше.
# Значение 'bars' отображает доверительный интервал в виде отрезков

sns.lineplot(x='year', y='passengers', err_style='bars', data=flights)
plt.show()
```

```python
# вернули прежний способ отображения доверительного интервала

sns.lineplot(x='year', y='passengers', err_style='band', data=flights)
plt.show()
```

```python
# If you set the err_style parameter to None, then the confidence interval will not be displayed
# Если установить для параметра err_style значение None, то доверительный интервал отображён не будет
```

```python
# Increasing the information content of the graph / Повышение информативности графика

# The lineplot() function provides three parameters that you can use to enhance the plot by associating additional set 
# features with visual aspects such as color, line style, and size. The parameters are responsible for this: 
# hue (color), style (line style) and size. Consider working with them with examples.

# Функция lineplot() предоставляет три параметра, используя которые можно повысить информативность графика, связав 
# дополнительные # признаки набора с такими визуальными аспектами как цвет, стиль линии и размер. 
# За это отвечают параметры: hue (цвет), style (стиль линии) и size (размер). Рассмотрим работу с ними на примерах.

import seaborn as sns
import pandas as pd
sns.set_style("darkgrid")
df = sns.load_dataset("mpg")

# Display the dependence of the car's power on the year of manufacture 
# Отобразим зависимость мощности автомобиля от года выпуска

sns.lineplot(x='model_year', y='horsepower', data=df) 
plt.show()
```

```python
# изменили ширину линии

sns.lineplot(x='model_year', y='horsepower', data=df, linewidth=.8) 
plt.show()
```

```python
# If the task is to study the dependence of car power on the year of manufacture separately for the USA and Japan, 
# then you can # use the hue parameter to highlight countries with color

# Если стоит задача изучить зависимость мощности автомобиля от года выпуска отдельно для стран США и Японии, то можно 
# воспользоваться параметром hue для выделения стран цветом

sns.lineplot(x='model_year', y='horsepower', hue='origin', data=df[df['origin'] != 'europe'])
plt.show()
```

```python
# изменили тип палитры

sns.lineplot(x='model_year', y='horsepower', hue='origin', data=df[df['origin'] != 'europe'], palette='pastel')
plt.show()
```

```python
#To further customize the color separation, you can use the following parameters of the lineplot() function: palette,
# hue_order, hue_norm

# Для дополнительной настройки цветового разделения можно использовать следующие параметры функции lineplot(): palette,
# hue_order, hue_norm

# Change the color and set the order / Изменим цвет и зададим порядок
df_usa_jp = df[df['origin'] != 'europe']
sns.lineplot(x='model_year', y='horsepower', hue='origin', 
palette={'usa':'r', 'japan':'b'}, hue_order=['japan', 'usa'], data=df_usa_jp)
plt.show()
```

```python
# добавили подпись

g = sns.lineplot(x='model_year', y='horsepower', hue='origin', 
palette={'usa':'r', 'japan':'b'}, hue_order=['japan', 'usa'], data=df_usa_jp)
g.set_title("Зависимость мощности автомобиля от года выпуска", fontsize=15)
plt.show()
```

```python
# To segment data using a line style, the style parameter is used, through which a feature is set, by which
# division is made

# Для сегментации данных с помощью стиля линии используется параметр style, через него задаётся признак, по которому будет 
# производиться разделение

sns.lineplot(x='model_year', y='horsepower', style='origin', data=df_usa_jp)
plt.show()
```

```python
# Visualizing relationships with a matte setting / Визуализация отношений с настройкой подложки

# The seaborn library provides another function for visualizing relationships in data: relplot(). 
# Unlike lineplot() and scatterplot(), it provides the ability to customize not only the appearance of the chart 
#itself, but also the figure - the substrate on which all graphic components are located

# Библиотека seaborn предоставляет ещё одну функцию для визуализации отношений в данных: relplot(). В отличие от lineplot() и
# scatterplot(), она предоставляет возможность настраивать не только внешний вид непосредственно самого графика, 
# но и фигуру — подложку, на которой располагаются все графические компоненты

sns.set_style('whitegrid')

iris = sns.load_dataset('iris')
sns.relplot(x='sepal_length', y='petal_length', hue='species', kind='scatter', data=iris)
plt.show()
```

```python
# поменяли размер фигуры

sns.relplot(x='sepal_length', y='petal_length', hue='species', kind='scatter', data=iris, height=3, aspect=7)
plt.show()
```

```python
# Let's build charts on three separate fields, as a separating parameter we will use the type of iris (species)
# Построим диаграммы на трёх отдельных полях, в качестве разделяющего параметра будем использовать тип ириса (species)

sns.relplot(x='sepal_length', y='petal_length', hue='species', kind='scatter', data=iris, col='species')
plt.show()
```

```python
# To improve the visibility of a chart that uses color separation through the hue parameter, you can 
# represent the resulting sets as visually distinguishable groups using the dodge parameter

# Для улучшения наглядности диаграммы, в которой используется цветовое разделение через параметр hue, можно
# представить полученные наборы в виде визуально различимых групп с помощью параметра dodge

sns.stripplot(x="time", y='tip', hue="sex", dodge=True, data=tips)
plt.show()
```

```python
# The swarmplot() function is similar to the stripplot() function in terms of its functionality and parameters for setting 
# the display, except for the jitter parameter, which it does not have. The idea of swarmplot() is that the displayed points 
# on the chart do not overlap, this allows you to draw conclusions about the predominance of certain values in data sets 
# by their visual distribution.

# Функция swarmplot() по своему функционалу и параметрам для настройки отображения подобна функции stripplot(), за исключением
# параметра jitter, который у неё отсутствует. Идея swarmplot() в том, что отображаемые точки на диаграмме не перекрываются, 
# это позволяет делать выводы о преобладании тех или иных значений в наборах данных по их визуальному распределению.

sns.swarmplot(x="time", y='tip', data=tips)
plt.show()
```

```python
# From the presented diagram, we can conclude that two dollars are most often left as a tip at lunch,
# and for dinner from two to four dollars 
# Из представленной диаграммы можно сделать вывод, что за ланчем чаще всего оставляют два доллара в качестве чаевых, 
# а за обедом от двух до четырёх долларов
```

```python
# Функция boxplot()

# Let's plot the mpg dataset, displaying the distribution of the number of miles a car travels on 
# one gallon of fuel for the US and Europe:
# Простроим диаграмму для набора данных mpg, отобразим распределение количества миль, которое автомобиль проезжает на
# одном галлоне топлива для США и Европы:

mpg = sns.load_dataset("mpg")
mpg_mod = mpg[mpg["origin"] != "japan"]
sns.boxplot(x="origin", y="mpg", data=mpg_mod)
plt.show()
```

```python
# Функция violin()

# In its functionality and capabilities, it is similar to the previously discussed boxplot (), 
# in addition, it displays an estimate of the density of the kernel.
# По своему функциональному назначению и возможностям она похожа на рассмотренную ранее boxplot(), 
# дополнительно на ней отображается оценка плотности ядра.

mpg_mod = mpg[mpg["origin"] != "japan"]
plt.figure(figsize=(10, 5))
plt.subplot(121)
sns.boxplot(x="origin", y="mpg", data=mpg_mod)
plt.subplot(122)
sns.violinplot(x="origin", y="mpg", data=mpg_mod)
plt.show()
```

```python
# ограничили диапазон violinplot в диапазоне наблюдаемых данных с помощью параметра cut

mpg_mod = mpg[mpg["origin"] != "japan"]
plt.figure(figsize=(10, 5))
plt.subplot(121)
sns.boxplot(x="origin", y="mpg", data=mpg_mod)
plt.subplot(122)
sns.violinplot(x="origin", y="mpg", data=mpg_mod, cut=0)
plt.show()
```

```python
from numpy import median
from numpy import mean

estimator = [mean, median, min, max]
plt.figure(figsize=(15, 5))
for i, es in enumerate(estimator):
 plt.subplot(1, len(estimator), i+1)
 plt.title(es.__name__)
 sns.barplot(x='origin', y='horsepower', estimator=es, data=mpg)
plt.show()
```

```python
# сделали "шапочки" у погрешностей
 
estimator = [mean, median, min, max]
plt.figure(figsize=(15, 5))
for i, es in enumerate(estimator):
 plt.subplot(1, len(estimator), i+1)
 plt.title(es.__name__)
 sns.barplot(x='origin', y='horsepower', estimator=es, data=mpg, capsize=.1)
plt.show()
```

```python
# Setting the confidence interval (or standard deviation) is done through the ci parameter:
# Задание доверительного интервала (или стандартного отклонения) производится через параметр ci:

cis = [None, 80, 'sd']
plt.figure(figsize=(15, 5))
for i, c in enumerate(cis):
 plt.subplot(1, len(cis), i+1)
 plt.title(f"ci = {c}")
 sns.barplot(x='origin', y='horsepower', ci=c, data=mpg)
plt.show()
```
